```python
import pandas as pd
import numpy as np
from numpy import nan
```

```python
pd.pandas.set_option('display.max_columns', None)
```

```python
train = pd.read_csv('final_train.csv')
train.head()
```

```python
test = pd.read_csv('final_test.csv')
test.head()
```

```python
y_train = train['SalePrice']
X_train = train.drop('SalePrice', axis=1)
X_test = test
X_test.shape, X_train.shape, y_train.shape
```

# Further Treating Outliers

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(contamination=0.1)
yhat = iso.fit_predict(X_train)
```

```python
mask = yhat != -1
X_train = X_train.iloc[mask,:]
y_train = y_train[mask]
```

```python
X_train.shape, y_train.shape
```

## Box plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
continuous = ['LotFrontage','LotArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','GarageArea',
              'WoodDeckSF','OpenPorchSF','EnclosedPorch','MasVnrArea','3SsnPorch','ScreenPorch','PoolArea','MiscVal']
for col in continuous:
    if 0 in X_train[col].unique():
        pass
    else:
        sns.boxplot(data=X_train, x=col)
        plt.ylabel(col)
        plt.title(str(col))
        plt.show()
```

# Feature Selection

```python
from sklearn.feature_selection import SelectKBest, mutual_info_regression
```

```python
#Select top 2 features based on mutual info regression
selector = SelectKBest(mutual_info_regression, k=20)
selector.fit(X_train, y_train)
X_train.columns[selector.get_support()]
```

```python
X_train = X_train[X_train.columns[selector.get_support()]]
X_train.head()
```

# Finding Best Parameters

```python
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
```

```python
reg = XGBRegressor()
reg.fit(X_train, y_train)
cv_results = cross_val_score(reg, X_train, y_train, scoring='neg_root_mean_squared_error', cv=5)
print(abs(cv_results))
print(abs(np.mean(cv_results)))
```

```python
n_estimators = [100,500,900,1100,1500]
max_depth = [2,3,5,10,15]
learning_rate = [0.05,0.1,0.15,0.20]
min_child_weight = [1,2,3,4]
booster = ['gbtree','gblinear']
base_score = [0.25,0.5,0.75,1]

param_grid = {'n_estimators':n_estimators,
              'max_depth':max_depth,
              'learning_rate':learning_rate,
              'min_child_weight':min_child_weight,
              'booster':booster,
              'base_score':base_score}
```

```python
random_cv = RandomizedSearchCV(estimator=reg, param_distributions=param_grid,
                    scoring='neg_root_mean_squared_error', return_train_score=True,
                    n_iter=100, cv=5, verbose=5, n_jobs=-1, random_state=13)
random_cv.fit(X_train, y_train)
```

```python
random_cv.best_estimator_, random_cv.best_score_
```

# Using Best Parameters

```python
reg = random_cv.best_estimator_
```

```python
reg.fit(X_train, y_train)
cv_results = cross_val_score(reg, X_train, y_train, scoring='neg_root_mean_squared_error', cv=5)
print(abs(cv_results))
print(abs(np.mean(cv_results)))
```

```python
y_pred = reg.predict(X_test[X_test.columns[selector.get_support()]])
y_pred
```

```python
from scipy.special import inv_boxcox
```

```python
final = inv_boxcox(y_pred, -0.07692391328663316)
final
```

```python
submission = pd.DataFrame(test['Id'])
submission['SalePrice'] = final
submission.head()
```

```python
submission.to_csv('submission.csv', index = False)
```
