```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from time import sleep
```

```python
file = 'event_details_ufc.csv'
df = pd.read_csv(file, sep=';')
```

```python
list_missing = []
df_already = pd.read_csv('fight_details_ufc.csv', sep=';')

for x in df['fight_link'].unique():
    if x not in df_already['fight_link'].unique():
        list_missing.append(x)

list_missing = df[(df['fight_link'].isin(list_missing)) & (df['strike']!='--')]['fight_link'].unique()
print(list_missing)
```

```python
# n_run = 0
# for n, link in enumerate(df['fight_link'].unique()[start_n:end_n]):
    # n_run += 1
    # if n_run > 1:
    #     break

for n, link in enumerate(list_missing):

    web = link
    # web = 'http://ufcstats.com/fight-details/e3aad51099a23ba4'
    result = requests.get(web)
    content = result.text

    soup = BeautifulSoup(content, 'lxml')

    # persons
    div_person = soup.find('div', class_='b-fight-details__persons clearfix')
    # details
    div_details = soup.find('div', class_='b-fight-details__fight')
    div_details_top = div_details.find('div', class_='b-fight-details__fight-head')
    div_details_bottom = div_details.find('div', class_='b-fight-details__content')
    # totals
    totals_table = soup.find_all('table', class_='b-fight-details__table js-fight-table')[0]
    # significant strikes
    sig_table = soup.find_all('table', class_='b-fight-details__table js-fight-table')[1]

    per_rd = totals_table.find_all('thead')
    totals_per_rd = totals_table.find('tbody')
    sig_per_rd = sig_table.find('tbody')

    for i, r in enumerate(per_rd):
        if i == 0:
            continue

        # fighter 1

        fight_link.append(link)
        win_loss.append(str.strip(div_person.find_all('div', class_='b-fight-details__person')[0].find('i').get_text(strip=True)))
        fighter.append(str.strip(div_person.find_all('div', class_='b-fight-details__person')[0].find('div', class_='b-fight-details__person-text').find('h3').find('a').get_text(strip=True)))
        fighter_link.append(str.strip(div_person.find_all('div', class_='b-fight-details__person')[0].find('div', class_='b-fight-details__person-text').find('h3').find('a')['href']))
        opponent.append(str.strip(div_person.find_all('div', class_='b-fight-details__person')[1].find('div', class_='b-fight-details__person-text').find('h3').find('a').get_text(strip=True)))
        opponent_link.append(str.strip(div_person.find_all('div', class_='b-fight-details__person')[1].find('div', class_='b-fight-details__person-text').find('h3').find('a')['href']))

        bout.append(str.strip(div_details_top.find('i', class_='b-fight-details__fight-title').get_text(strip=True)))
        method.append(str.strip(div_details_bottom.find_all('p')[0].find_all('i', class_='b-fight-details__text-item_first')[0].find_all('i')[1].get_text(" ", strip=True)))
        rd_end.append(str.replace(div_details_bottom.find_all('p')[0].find_all('i', class_='b-fight-details__text-item')[0].get_text(" ", strip=True), 'Round: ', ''))
        rd_time.append(str.replace(div_details_bottom.find_all('p')[0].find_all('i', class_='b-fight-details__text-item')[1].get_text(" ", strip=True), 'Time: ', ''))
        rd_format.append(str.replace(div_details_bottom.find_all('p')[0].find_all('i', class_='b-fight-details__text-item')[2].get_text(" ", strip=True), 'Time format: ', ''))
        referee.append(str.replace(div_details_bottom.find_all('p')[0].find_all('i', class_='b-fight-details__text-item')[3].get_text(" ", strip=True), 'Referee: ', ''))
        details.append(str.replace(div_details_bottom.find_all('p')[1].get_text(" ", strip=True), 'Details: ', ''))

        rd_num.append(r.get_text(strip=True))

        # fighter 2
        
        fight_link.append(link)
        win_loss.append(str.strip(div_person.find_all('div', class_='b-fight-details__person')[1].find('i').get_text(strip=True)))
        fighter.append(str.strip(div_person.find_all('div', class_='b-fight-details__person')[1].find('div', class_='b-fight-details__person-text').find('h3').find('a').get_text(strip=True)))
        fighter_link.append(str.strip(div_person.find_all('div', class_='b-fight-details__person')[1].find('div', class_='b-fight-details__person-text').find('h3').find('a')['href']))
        opponent.append(str.strip(div_person.find_all('div', class_='b-fight-details__person')[0].find('div', class_='b-fight-details__person-text').find('h3').find('a').get_text(strip=True)))
        opponent_link.append(str.strip(div_person.find_all('div', class_='b-fight-details__person')[0].find('div', class_='b-fight-details__person-text').find('h3').find('a')['href']))

        bout.append(str.strip(div_details_top.find('i', class_='b-fight-details__fight-title').get_text(strip=True)))
        method.append(str.strip(div_details_bottom.find_all('p')[0].find_all('i', class_='b-fight-details__text-item_first')[0].find_all('i')[1].get_text(" ", strip=True)))
        rd_end.append(str.replace(div_details_bottom.find_all('p')[0].find_all('i', class_='b-fight-details__text-item')[0].get_text(" ", strip=True), 'Round: ', ''))
        rd_time.append(str.replace(div_details_bottom.find_all('p')[0].find_all('i', class_='b-fight-details__text-item')[1].get_text(" ", strip=True), 'Time: ', ''))
        rd_format.append(str.replace(div_details_bottom.find_all('p')[0].find_all('i', class_='b-fight-details__text-item')[2].get_text(" ", strip=True), 'Time format: ', ''))
        referee.append(str.replace(div_details_bottom.find_all('p')[0].find_all('i', class_='b-fight-details__text-item')[3].get_text(" ", strip=True), 'Referee: ', ''))
        details.append(str.replace(div_details_bottom.find_all('p')[1].get_text(" ", strip=True), 'Details: ', ''))

        rd_num.append(r.get_text(strip=True))


    for i in range(len(per_rd)-1):

        # fighter 1

        kd.append(totals_per_rd.find_all('tr')[i].find_all('td')[1].find_all('p')[0].get_text(strip=True))
        sig_str_landed.append(str.split(totals_per_rd.find_all('tr')[i].find_all('td')[2].find_all('p')[0].get_text(strip=True), sep=' of ')[0])
        sig_str_total.append(str.split(totals_per_rd.find_all('tr')[i].find_all('td')[2].find_all('p')[0].get_text(strip=True), sep=' of ')[1])
        sig_str_pct.append(totals_per_rd.find_all('tr')[i].find_all('td')[3].find_all('p')[0].get_text(strip=True))
        total_str_landed.append(str.split(totals_per_rd.find_all('tr')[i].find_all('td')[4].find_all('p')[0].get_text(strip=True), sep=' of ')[0])
        total_str_total.append(str.split(totals_per_rd.find_all('tr')[i].find_all('td')[4].find_all('p')[0].get_text(strip=True), sep=' of ')[1])
        tdown_landed.append(str.split(totals_per_rd.find_all('tr')[i].find_all('td')[5].find_all('p')[0].get_text(strip=True), sep=' of ')[0])
        tdown_total.append(str.split(totals_per_rd.find_all('tr')[i].find_all('td')[5].find_all('p')[0].get_text(strip=True), sep=' of ')[1])
        tdown_pct.append(totals_per_rd.find_all('tr')[i].find_all('td')[6].find_all('p')[0].get_text(strip=True))
        sub_att.append(totals_per_rd.find_all('tr')[i].find_all('td')[7].find_all('p')[0].get_text(strip=True))
        rev.append(totals_per_rd.find_all('tr')[i].find_all('td')[8].find_all('p')[0].get_text(strip=True))

        sig_head_landed.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[3].find_all('p')[0].get_text(strip=True), sep=' of ')[0])
        sig_head_total.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[3].find_all('p')[0].get_text(strip=True), sep=' of ')[1])
        sig_body_landed.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[4].find_all('p')[0].get_text(strip=True), sep=' of ')[0])
        sig_body_total.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[4].find_all('p')[0].get_text(strip=True), sep=' of ')[1])
        sig_leg_landed.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[5].find_all('p')[0].get_text(strip=True), sep=' of ')[0])
        sig_leg_total.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[5].find_all('p')[0].get_text(strip=True), sep=' of ')[1])
        sig_distance_landed.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[6].find_all('p')[0].get_text(strip=True), sep=' of ')[0])
        sig_distance_total.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[6].find_all('p')[0].get_text(strip=True), sep=' of ')[1])
        sig_clinch_landed.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[7].find_all('p')[0].get_text(strip=True), sep=' of ')[0])
        sig_clinch_total.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[7].find_all('p')[0].get_text(strip=True), sep=' of ')[1])
        sig_ground_landed.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[8].find_all('p')[0].get_text(strip=True), sep=' of ')[0])
        sig_ground_total.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[8].find_all('p')[0].get_text(strip=True), sep=' of ')[1])
        
        # fighter 2

        kd.append(totals_per_rd.find_all('tr')[i].find_all('td')[1].find_all('p')[1].get_text(strip=True))
        sig_str_landed.append(str.split(totals_per_rd.find_all('tr')[i].find_all('td')[2].find_all('p')[1].get_text(strip=True), sep=' of ')[0])
        sig_str_total.append(str.split(totals_per_rd.find_all('tr')[i].find_all('td')[2].find_all('p')[1].get_text(strip=True), sep=' of ')[1])
        sig_str_pct.append(totals_per_rd.find_all('tr')[i].find_all('td')[3].find_all('p')[1].get_text(strip=True))
        total_str_landed.append(str.split(totals_per_rd.find_all('tr')[i].find_all('td')[4].find_all('p')[1].get_text(strip=True), sep=' of ')[0])
        total_str_total.append(str.split(totals_per_rd.find_all('tr')[i].find_all('td')[4].find_all('p')[1].get_text(strip=True), sep=' of ')[1])
        tdown_landed.append(str.split(totals_per_rd.find_all('tr')[i].find_all('td')[5].find_all('p')[1].get_text(strip=True), sep=' of ')[0])
        tdown_total.append(str.split(totals_per_rd.find_all('tr')[i].find_all('td')[5].find_all('p')[1].get_text(strip=True), sep=' of ')[1])
        tdown_pct.append(totals_per_rd.find_all('tr')[i].find_all('td')[6].find_all('p')[1].get_text(strip=True))
        sub_att.append(totals_per_rd.find_all('tr')[i].find_all('td')[7].find_all('p')[1].get_text(strip=True))
        rev.append(totals_per_rd.find_all('tr')[i].find_all('td')[8].find_all('p')[1].get_text(strip=True))

        sig_head_landed.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[3].find_all('p')[1].get_text(strip=True), sep=' of ')[0])
        sig_head_total.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[3].find_all('p')[1].get_text(strip=True), sep=' of ')[1])
        sig_body_landed.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[4].find_all('p')[1].get_text(strip=True), sep=' of ')[0])
        sig_body_total.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[4].find_all('p')[1].get_text(strip=True), sep=' of ')[1])
        sig_leg_landed.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[5].find_all('p')[1].get_text(strip=True), sep=' of ')[0])
        sig_leg_total.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[5].find_all('p')[1].get_text(strip=True), sep=' of ')[1])
        sig_distance_landed.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[6].find_all('p')[1].get_text(strip=True), sep=' of ')[0])
        sig_distance_total.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[6].find_all('p')[1].get_text(strip=True), sep=' of ')[1])
        sig_clinch_landed.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[7].find_all('p')[1].get_text(strip=True), sep=' of ')[0])
        sig_clinch_total.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[7].find_all('p')[1].get_text(strip=True), sep=' of ')[1])
        sig_ground_landed.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[8].find_all('p')[1].get_text(strip=True), sep=' of ')[0])
        sig_ground_total.append(str.split(sig_per_rd.find_all('tr')[i].find_all('td')[8].find_all('p')[1].get_text(strip=True), sep=' of ')[1])        

    print("successful row #", n+1)
    # delay        
    sleep(2)
```
