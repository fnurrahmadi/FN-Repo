```python
import requests
import json
import pandas as pd
from time import sleep

pd.set_option('display.max_columns', None)
```

```python
df_final = pd.DataFrame()
df_already = pd.read_csv('all_anime.csv', sep=';')
```

```python
start_id = 55000
# end_id = 55348 55517
end_id = 55517
for id in range(start_id, end_id):
    id += 1
    
    link = f'https://api.jikan.moe/v4/anime/{id}/full'

    result = requests.get(link)
    sleep(1.2)
    
    if result.status_code == 200:
        df = pd.DataFrame.from_dict(result.json(), orient='index')
        df_final = pd.concat([df_final, df])
        print('done', id)
    else:
        print(result.status_code, id)
        continue
```

```python
df_final = pd.concat([df_already, df_final])
```

```python
df_final.tail()
```

```python
df_final.to_csv('all_anime.csv', index=False, encoding='utf-8', sep=';')

print('done', len(df_final['url'].unique()), '\n\nsave successful!')
```
