```python
import pandas as pd
```

```python
df_movies = pd.read_csv('scraped_transcriptions.csv', encoding='utf-8', delimiter=';')
df_movies.head()
```

```python
df_scripts = pd.read_csv('scrape_transcriptions_1.csv', encoding='utf-8', delimiter=';')
df_scripts.head()
```

```python
df = pd.merge(df_movies, df_scripts, how='right', on='links').drop_duplicates('links')
df.info()
```

```python
df[df['script'].isnull()].drop_duplicates()
```

```python
df[df['script'].isnull()].drop_duplicates().to_csv('null_scripts.csv', index=False, encoding='utf-8', sep=';')
```

```python
dataset = pd.merge(df_movies, df_scripts, how='inner', on='links')
dataset.info()
```

```python
dataset = dataset.drop_duplicates('links')[~dataset['script'].isnull()]
dataset.info()
```

```python
dataset.to_csv('dataset.csv', index=False, encoding='utf-8', sep=';')
```
